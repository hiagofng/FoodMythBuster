```python
import duckdb
conn = duckdb.connect("../data/foodmythbuster.duckdb")
```

```python
conn.sql("""
SELECT
    brands,
    count(*)                                                AS total,
    sum(is_deceptive::INT)                                  AS deceptive,
    round(avg(is_deceptive::INT) * 100, 1)                  AS deceptive_pct
FROM staging.stg_products
WHERE brands IS NOT NULL
GROUP BY brands
HAVING count(*) >= 5
ORDER BY deceptive_pct DESC
LIMIT 20;
""").show()
```

```python
conn.sql("""
SELECT
*
FROM staging.stg_products
LIMIT 10;
""").show()
```
